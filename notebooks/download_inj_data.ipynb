## Scrape all movement to and from IL and all games missed due to injury from Pro Sports Transactions

```python
from urllib.request import urlopen
from urllib.error import HTTPError
from bs4 import BeautifulSoup, Comment
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import datetime
import unidecode

from googlesearch import search

mpl.rcParams['font.family'] = "serif"
mpl.rcParams['text.usetex'] = True
mpl.rcParams['ytick.right'] = True
mpl.rcParams['ytick.direction'] = "in"
mpl.rcParams['xtick.top'] = True
mpl.rcParams['xtick.direction'] = "in"
mpl.rcParams['axes.titlesize'] = 20
mpl.rcParams['axes.labelsize'] = 20
mpl.rcParams['xtick.labelsize'] = 16
mpl.rcParams['ytick.labelsize'] = 16
mpl.rcParams['legend.fontsize'] = 20
mpl.rcParams['figure.dpi'] = 100
mpl.rcParams['savefig.dpi'] = 100
mpl.rcParams['figure.figsize'] = (6,4)
plt.rcParams['axes.labelweight'] = 'bold'
mpl.rcParams['text.latex.preamble']=[r"\usepackage{amsmath}"]    
```

### Loop through all pages on Pro Sports Transactions for the Query of all movement to IL and missed games due to injury

```python
headers = ['Date', 'Team', 'Acquired', 'Relinquished', 'Notes']
N_pages = 1067 #Number of pages the query will produce 

for i in range(N_pages): #Number of results from prosports transactions
    
    print("Currently on page {}/{}".format(i+1,N_pages))

    url = "https://www.prosportstransactions.com/basketball/Search/SearchResults.php?Player=&Team=&BeginDate=2010-10-01&EndDate=&ILChkBx=yes&InjuriesChkBx=yes&Submit=Search&start={}".format(25*i)

    html = urlopen(url)
    soup = BeautifulSoup(html)
    
    body = soup.findAll('td')
    
    Ncols = len(headers)
    Nrows = int(len(body)/len(headers))-2

    vals_tmp = np.zeros((Nrows,Ncols),dtype=object)
    
    for j in range(1,Nrows):
        for k in range(Ncols):
            val = body[j*Ncols+k].getText()
            if k==0:
                slice_ind = 0
            elif k==2:
                slice_ind = 3
            else:
                slice_ind = 1

            vals_tmp[j-1,k] = val[slice_ind:]
            
    if i==0:
        values = vals_tmp
    else:
        values = np.concatenate((values,vals_tmp))
```

```python
full_inj_df = pd.DataFrame(values,columns=headers)
full_inj_df.head(10)
```

### Data has been downloaded! Now clean up a bit, in particular removing all "injuries" that were the result of an illness or something else we don't want to include

```python
full_inj_df = full_inj_df.replace(to_replace=0,value=None)

full_inj_df = full_inj_df.loc[full_inj_df['Acquired'] == '']
full_inj_df.drop_duplicates(subset=['Relinquished','Date'],inplace=True)
full_inj_df = full_inj_df.loc[~full_inj_df['Notes'].isnull()].reset_index(drop=True)
full_inj_df = full_inj_df[full_inj_df['Notes'] != '0']

full_inj_df = full_inj_df[~full_inj_df['Notes'].str.contains('surgery|gastro|flu|illness|virus|declined\
|rest|sinus|strep|concussion|infection|upset|suspension')].reset_index()

player_names = full_inj_df.Relinquished.unique()

print("Found {} injuries listed for {} players".format(full_inj_df.shape[0],len(player_names)))
```

#### Cut the leading dots from player names

```python
cut = lambda s: s[2:]
full_inj_df['Relinquished'] = full_inj_df['Relinquished'].apply(cut)
```

### Cut out player names with slashes

```python
name_dict = {}
slashes = full_inj_df[full_inj_df['Relinquished'].str.contains('/')]
names = slashes.Relinquished.unique()

for name_slash in names:
    name = name_slash.split('/')[0]
    name = name.strip()

    name_dict[name_slash] = name
```

```python
def fix_slashes(name):
    if name in name_dict:
        return name_dict[name]
    else:
        return name
```

```python
full_inj_df['Relinquished'] = full_inj_df['Relinquished'].apply(fix_slashes)
```

#### Get our final count

```python
player_names = full_inj_df.Relinquished.unique()

print("Found {} injuries listed for {} players".format(full_inj_df.shape[0],len(player_names)))
```
