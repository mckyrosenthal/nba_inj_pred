## This notebook is used to scrape speed and distance tracking data from stats.nba.com. Careful with running this, as they will sometimes reject or slow down requests if too many come in.

```python
import pandas as pd
import numpy as np
import datetime

def str2datetime(date):
    
    date_tmp = date.split('-')
    
    y = int(date_tmp[0])
    m = int(date_tmp[1])
    d = int(date_tmp[2])
    
    return datetime.datetime(y,m,d)
```

### We need to classify every games as a playoff or a regular season game to pull the speed and distanace tracking stats. Start by defining a dictionary of when the playoffs started every year.

```python
full_df = pd.read_csv('nba_games_2010_20.csv')
full_df = full_df.loc[full_df['Age'] != 0]
full_df = full_df.drop(columns='Unnamed: 0')
print(full_df.shape)
```

```python
playoff_date = {'2011':datetime.datetime(2011,4,16),\
                '2012':datetime.datetime(2012,4,28),\
               '2013':datetime.datetime(2013,4,20),
                '2014':datetime.datetime(2014,4,19),
                '2015':datetime.datetime(2015,4,18),
                '2016':datetime.datetime(2016,4,16),
                '2017':datetime.datetime(2017,4,15),
                '2018':datetime.datetime(2018,4,14),
                '2019':datetime.datetime(2019,4,13),
                '2020':datetime.datetime(2020,4,18),
               }
```

```python
def reg_or_play(row):
    date = row['Date']
    date = str2datetime(date)
    s_end = datetime.datetime(date.year,7,1)
    if date > s_end:
        season = str(date.year + 1)
    else:
        season = str(date.year)
    
    if playoff_date[season] < date:
        return "Play"
    else:
        return "Reg"
```

```python
gm_type = full_df.apply(reg_or_play,axis=1)
full_df['gm_type'] = gm_type
```

```python
reg_s = full_df.loc[full_df['gm_type']=="Reg"]
```

```python
reg_s = reg_s.sort_values(by=['Date','Name']).reset_index(drop=True)
```

```python
reg_s
```

### NBA doesn't have any speed/distance tracking before the 2013-14 season, so drop all rows before that period of time

```python
to_drop = []

for i,r in reg_s.iterrows():
    if str2datetime(r['Date'])<datetime.datetime(2013,8,1):
        to_drop.append(i)
    
    if i%1000==0:
        print(i)
```

```python
reg_s.drop(to_drop,inplace=True)
```

```python
reg_s.reset_index(drop=True,inplace=True)
```

### Loop over regular season dates, and for each one pull the average stats from a specified time window

#### Now we're going to use selenium to open up stats.nba.com. Some of these attempts will fail, either because we're looking for stats from before the season started (because we pull average stats from the day *before* the game), or just because the webpage will hang (they're quick to reject requests).

```python
from selenium.common.exceptions import NoSuchElementException
from selenium import webdriver
import time
```

```python
dates = reg_s.Date.unique()
time_window = datetime.timedelta(days=21) #IMPORTANT! The time window over which the stats will be pulled from

vals = []
dates_failed = []
path_to_chromedriver = '/Users/michaelrosenthal/Documents/chromedriver' # Path to access a chrome driver
browser = webdriver.Chrome(executable_path=path_to_chromedriver)

for date in dates[266:]:
    
    date_tmp = date.split('-')   
    y = int(date_tmp[0])
    m = int(date_tmp[1])
    d = int(date_tmp[2])
    
    date_datetime = datetime.datetime(y,m,d)
    season = y if date_datetime<datetime.datetime(y,8,1) else y+1
    season_str = "{}-{}".format(season-1,season-2000)
    date_to = date_datetime - datetime.timedelta(days=1)
    date_from = date_to - time_window
    
    url = "https://stats.nba.com/players/speed-distance/?Season={}&SeasonType=Regular%20Season&DateFrom={}%2F{}%2F{}&DateTo={}%2F{}%2F{}"\
    .format(season_str,date_from.month,date_from.day,date_from.year,date_to.month,date_to.day,date_to.year)
    
    browser.get(url)
    time.sleep(0.5)
    path = '/html/body/main/div[2]/div/div[2]/div/div/nba-stat-table/div[3]/div/div/select/option[1]'
    
    try:
        browser.find_element_by_xpath(path).click() #Try clicking on the button to display all stats at once
        
        table = browser.find_element_by_class_name('nba-stat-table__overflow')
        data = table.text.split('\n')

        for i in range(1,int((len(data)-1)/2)+1):
            val_0 = np.array([date])
            val_1 = np.array([data[2*i-1]])
            val_2= np.array(data[2*i].split(' '))
            val_tmp = np.concatenate((val_0,val_1,val_2))

            vals.append(val_tmp)
        
    except NoSuchElementException:
        dates_failed.append(date)
        time.sleep(0.5)
        continue
        
    
browser.quit()
```
